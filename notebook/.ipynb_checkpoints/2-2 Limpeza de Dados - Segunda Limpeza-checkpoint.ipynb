### Tratamento de Erros do Scrapping

Por algum motivo os valores de preço vieram com erros a partir do Scrapping. O era para ser 1022.0 está 1.022.

Neste caso, vamos fazer a correção desses valores.

As variáveis que estão com este problema são: **Preço/Noite** e **Preço com taxas**

```python
print(df_raw['Preço/Noite'].sort_values())
```

```python
# Preço/Noite:
print(len(df_raw.loc[df_raw['Preço/Noite'] < 10.0]))
```

```python
# Preço com taxas
print(len(df_raw.loc[df_raw['Preço com taxas'] < 10.0]))
```

### Preço/Noite

```python
filter_1 = df_raw.loc[(df_raw['Preço/Noite'] < 10.0), :].index
```

```python
filter_1
```

```python
df_raw.iloc[filter_1]['Preço/Noite'] * 1000
```

```python
df_raw['Novo_Preço/Noite'] = df_raw.iloc[filter_1]['Preço/Noite'] * 1000
```

```python
values = df_raw['Preço/Noite']
```

```python
df_raw['Novo_Preço/Noite'] = df_raw['Novo_Preço/Noite'].fillna(values)
```

```python
df_raw.head()
```

**Verificação:**

```python
# Preço/Noite:
print(len(df_raw.loc[df_raw['Novo_Preço/Noite'] < 10.0]))
```

### Preço com taxas

```python
filter_2 = df_raw.loc[(df_raw['Preço com taxas'] < 10.0), :].index
```

```python
filter_2
```

```python
df_raw['Novo_Preço com taxas'] = df_raw.iloc[filter_2]['Preço com taxas'] * 1000
```

```python
value_2 = df_raw['Preço com taxas']
```

```python
df_raw['Novo_Preço com taxas'] = df_raw['Novo_Preço com taxas'].fillna(value_2)
```

```python
df_raw.head()
```

**Verificação**

```python
# Preço com taxas:
df_raw.loc[df_raw['Novo_Preço com taxas'] < 10.0]
```

Os erros foram tratados.

Agora vamos dropar as variáveis com erro  renomea-las.

```python
df_raw = df_raw.drop({'Preço/Noite', 'Preço com taxas'}, axis=1)
```

```python
df_raw.rename(columns = {'Novo_Preço/Noite': 'Preço/Noite', 
                                      'Novo_Preço com taxas': 'Preço com taxas'}, inplace=True)
```

## Tratamento finalizado

# 2.2.3 Função de Tratamento

```python
df_test_funccion = pd.read_csv('baixada_santista.csv')
```

```python
def corrigir_monetario(df, name='name'):
    """
    Correção de valores float incorretos 1.0 para 1000.0
    df = Dataframe
    name= Nome da variável    
    """
    
    filter_1 = df.loc[(df[name] < 10.0), :].index   
    df['New_name'] = df.iloc[filter_1][name] * 1000    
    value = df[name]    
    df['New_name'] = df['New_name'].fillna(value)
    df = df.drop(name, axis=1)
    df.rename(columns = {'New_name': name}, inplace=True)
    
    return df
```

```python
df_test_funccion = corrigir_monetario(df_test_funccion, name='Preço/Noite')
```

```python
df_test_funccion = corrigir_monetario(df_test_funccion, name='Preço com taxas')
```
